```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('/content/engineered_data (3).csv')
print(df.head())

#Calculate survival rate by different groups (e.g., Pclass, Sex)
survival_bu_class = df.groupby('Pclass')['Survived'].mean()
print(survival_bu_class)

#Calculate summary statistics for numerical features
print(df[['Age','Fare']].describe())


#Bar Chart: Survival count by Sex
# Use 'Sex_male' instead of 'Sex' because the original column was transformed during encoding
sns.countplot(x='Survived',hue='Sex_male',data=df , color='pink')
plt.title('Survival Count by Sex(Male=1 , Faemale=0)')
plt.show()


#Identify and select only numerical columns for analysis
numerical_df = df.select_dtypes(include=[np.number])

#
correlation_matrix = numerical_df.corr()

#Heatmap: Correlation matrix
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix,annot=True,cmap='RdPu')
plt.title('Correlation Analysis of Numerical Features')
plt.show()




#Box Plot: Fare distribution by Survival
sns.boxplot(x='Survived',y='Fare',data=df , color='purple')
plt.title('Fare Distribution by Survival')
plt.show()


```
