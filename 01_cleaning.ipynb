```python
from pandas._libs import missing
from sklearn.base import re
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder


df = pd.read_csv('/content/train.csv')
print(df.head())
print(df.shape)
print(df.info())

def clean_types(data):
  """Label Encoding was applied to the (Sex) column to convert categorical values into numerical form.
  In addition, One-Hot Encoding was performed on the categorical columns (Embarked) and (Cabin) using the get_dummies function,
   with the first category dropped (drop_first=True) to avoid multicollinearity.
   These preprocessing steps aim to prepare the dataset in a numerical format suitable for analysis and modeling."""
  df1=data.copy()
  le = LabelEncoder()
  df1['Sex'] = le.fit_transform(df1['Sex'])
  # categorical_cols = ['Embarked', 'Cabin']
  # data = pd.get_dummies(df1, columns=categorical_cols, drop_first=True)
  return data


def handle_missing(data):
  df1=data.copy()
  """A copy of the original dataset was created to avoid modifying the original data.
  The number of missing values in each column was then computed using the isnull().sum()
   function to display the data status before preprocessing.
   The analysis revealed missing values in the Age, Embarked, and Cabin columns"""
  missing_values = df1.isnull().sum()
  print("Missing values before filling:\n",missing_values)
  """Missing values in the Age column were handled using mean imputation.
  Rows with missing values in the Embarked column were removed due to their small number"""
  df1['Age']= df1['Age'].fillna(df1['Age'].mean())
  df1= df1.dropna(subset=['Embarked'])
  """For the Cabin column, missing values were replaced with "Unknown" to represent unavailable data"""
  df1['Cabin']= df1['Cabin'].fillna('Unknown')
  """Finally, missing values were recalculated to verify that the preprocessing steps were successfully applied"""
  missing_values1 = df1.isnull().sum()
  print("Missing values after filling:\n",missing_values1)
  return df1

def handle_duplicates(data):
  """A copy of the original dataset was created to avoid modifying the original data.
   A check for duplicate records was then performed using the duplicated().sum() function to count the number of duplicates.
   The results showed that there were no duplicate entries; therefore, no duplicate removal step was applied, and the dataset was retained as is"""
  df1=data.copy()
  duplicates = df1.duplicated().sum()
  print("Number of duplicates before removing:", duplicates)
  return df1

def handle_outliers(data):
   """A copy of the original dataset was created to avoid direct modification.
   Outliers in the numerical columns (Age) and (Fare) were handled using the Interquartile Range (IQR) method"""
   df1=data.copy()
   for col in ['Age', 'Fare']:
    """The first quartile (Q1) and third quartile (Q3) were calculated,
    followed by the computation of the IQR (Q3 − Q1).
     Lower and upper bounds were then defined using the formula:
     Q1 − 1.5 × IQR and Q3 + 1.5 × IQR"""
    q1 = df1[col].quantile(0.25)
    q3 = df1[col].quantile(0.75)
    IQR = q3 - q1

    lower = q1 - 1.5 * IQR
    upper = q3 + 1.5 * IQR

    """Finally, the clip function was applied to cap values outside these bounds,
    replacing outliers with the nearest acceptable limit"""
    df1[col] = df1[col].clip(lower, upper)
   return df1

def clean_data(data):
  """A function named clean_data was defined to sequentially apply all preprocessing steps to the dataset.
   First, a copy of the original data was created to prevent direct modification.
    Then, functions for handling missing values (handle_missing),
    data type transformation (clean_types),
     duplicate checking (handle_duplicates),
      and outlier treatment (handle_outliers) were applied
      This function aims to streamline and standardize the data cleaning process,
      preparing the dataset for subsequent analysis and modeling stages"""
  df1= data.copy()
  df1 = handle_missing(df1)
  df1 = clean_types(df1)
  df1 = handle_duplicates(df1)
  df1 = handle_outliers(df1)
  return df1
clean_data(df)

df.to_csv('cleaned_data.csv', index=False)

from google.colab import files
files.download('cleaned_data.csv')
```
